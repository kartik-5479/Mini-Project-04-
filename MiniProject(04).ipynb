# **ComboFinder**
---
####**Problem Statement:**
####Find all the list of products whose sum-of-price is between 290 and 310.
#### ProductList = {'p1':10, 'p2':15, 'p3':20, 'p4':25, 'p5':30, 'p6':35,'p7':50,'p8':40, 'p9':55, 'p10':60, 'p11':65, 'p12':75, 'p13':70,'p14':45}
---

```python
from itertools import combinations

# Product dictionary
ProductList = {
    'p1': 10, 'p2': 15, 'p3': 20, 'p4': 25, 'p5': 30, 'p6': 35,
    'p7': 50, 'p8': 40, 'p9': 55, 'p10': 60, 'p11': 65, 'p12': 75,
    'p13': 70, 'p14': 45
}

# Range of valid total prices
min_total = 290
max_total = 310

# Store valid combinations
valid_combinations = []

# Check all possible combinations (length 2 to total products)
product_items = list(ProductList.items())

for r in range(2, len(ProductList)+1):
    for combo in combinations(product_items, r):
        combo_names = [name for name, _ in combo]
        combo_price = sum(price for _, price in combo)
        if min_total <= combo_price <= max_total:
            valid_combinations.append((combo_names, combo_price))

# Print valid combinations
for names, total in valid_combinations:
    print(f"Products: {names} -> Total: ₹{total}")
```
